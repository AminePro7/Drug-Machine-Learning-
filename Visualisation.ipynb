```python
#Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Reading Data Set
df_drug = pd.read_csv("drug200.csv")
df_drug.head()
```

```python
#Representation des variables catégorielles en fct de leurs nombre 
```

```python
#La ditrubution des médicaments selon leurs types 
sns.set_theme(style="darkgrid")
sns.countplot(y="Drug", data=df_drug, palette="flare")
plt.ylabel('Drug Type')
plt.xlabel('Total')
plt.show()
```

```python
#La distribution des personnes selon le genre 
sns.set_theme(style="darkgrid")
sns.countplot(x="Sex", data=df_drug, palette="rocket")
plt.xlabel('Gender (F=Female, M=Male)')
plt.ylabel('Total')
plt.show()
```

```python
#La distribution selon la pression du sang
sns.set_theme(style="darkgrid")
sns.countplot(y="BP", data=df_drug, palette="crest")
plt.ylabel('Blood Pressure')
plt.xlabel('Total')
plt.show()
```

```python
#La distribution de Cholesterol 
sns.set_theme(style="darkgrid")
sns.countplot(x="Cholesterol", data=df_drug, palette="magma")
plt.xlabel('Blood Pressure')
plt.ylabel('Total')
plt.show()
```

```python
#Representation des variables categorielles en fonction d'une autre variable categorielle
```

```python
#Les medicaments et leurs consommation selon le genre 
pd.crosstab(df_drug.Sex,df_drug.Drug).plot(kind="bar",figsize=(12,5),color=['#003f5c','#ffa600','#58508d','#bc5090','#ff6361'])
plt.title('Gender distribution based on Drug type')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.ylabel('Frequency')
plt.show()
```

```python
#La pression du sang selon le Cholesetrol 
pd.crosstab(df_drug.BP,df_drug.Cholesterol).plot(kind="bar",figsize=(15,6),color=['#6929c4','#1192e8'])
plt.title('Blood Pressure distribution based on Cholesterol')
plt.xlabel('Blood Pressure')
plt.xticks(rotation=0)
plt.ylabel('Frequency')
plt.show()
```

```python
#Les médicaments selon la pression du sang 
df_BP_Drug = df_drug.groupby(["Drug","BP"]).size().reset_index(name = "Count")
plt.figure(figsize = (9,5))
sns.barplot(x = "Drug",y="Count", hue = "BP",data = df_BP_Drug)
plt.title("BP -- Drug")
plt.show()
```

```python
#La consommation des medicaments selon les niveaux de Cholesterol
df_CH_Drug = df_drug.groupby(["Drug","Cholesterol"]).size().reset_index(name = "Count")
plt.figure(figsize = (9,5))
sns.barplot(x = "Drug",y="Count", hue = "Cholesterol",data = df_CH_Drug)
plt.title("Cholesterol -- Drug")
plt.show()
```

```python
#La distribution des variables numeriques filtré en fonction d'une variable catégorielle
```

```python
#La distribution de Na_to_K en fonction de l'age puis les points resultats sont filtré selon le genre 
plt.scatter(x=df_drug.Age[df_drug.Sex=='F'], y=df_drug.Na_to_K[(df_drug.Sex=='F')], c="Blue")
plt.scatter(x=df_drug.Age[df_drug.Sex=='M'], y=df_drug.Na_to_K[(df_drug.Sex=='M')], c="Orange")
plt.legend(["Female", "Male"])
plt.xlabel("Age")
plt.ylabel("Na_to_K")
plt.show()
```

```python
#Representation d'une variable numérique en fonction d'une variable categorielle
```

```python
#La consommation des médicaments selon l'age 
plt.figure(figsize = (9,5))
sns.swarmplot(x = "Drug", y = "Age",data = df_drug)
plt.legend(df_drug.Drug.value_counts().index)
plt.title("Age -- Drug")
plt.show()
```

```python
#La consommation des médicaments selon Na_to_K
plt.figure(figsize = (9,5))
sns.swarmplot(x = "Drug", y = "Na_to_K",data = df_drug)
plt.title("Na_to_K -- Drug")
plt.show()
```

```python
#La consommation des médicaments de Na_to_K puis on les filtre selon la pression du sang 
plt.figure(figsize = (9,5))
sns.swarmplot(x = "Drug", y = "Na_to_K",hue="BP",data = df_drug)
plt.legend()
plt.title("Na_to_K -- BP -- Drug")
plt.show()
```
